# Grover Amplitude Visualization

This notebook visualizes how amplitudes evolve during the algorithm:
- before the oracle
- after phase marking
- after diffusion

Seeing the amplitudes directly makes it easier to understand how interference amplifies the correct answer.

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

# create circuit
qc = QuantumCircuit(2)

# step 1: equal superposition

qc.h(0)
qc.h(1)

print("state after superposition:\n")

state = Statevector.from_instruction(qc)

print(state)

# step 2: oracle marks |11>

qc.cz(0,1)

print("\nstate after oracle:\n")

state = Statevector.from_instruction(qc)

print(state)

# step 3: diffusion operator

qc.h(0)
qc.h(1)

qc.x(0)
qc.x(1)

qc.cz(0,1)

qc.x(0)
qc.x(1)

qc.h(0)
qc.h(1)

print("\nstate after diffusion:\n")

state = Statevector.from_instruction(qc)

print(state)
```

# Observation

## 1. After Superposition

[0.5, 0.5, 0.5, 0.5]

This corresponded to:

|00>
|01>
|10>
|11>

all having:

equal amplitude.

Since probability is:

|amplitude|^2

each probability became:

(0.5)^2 = 0.25

So initially every state had:

25%

probability.

At this stage the search space was completely uniform.

The algorithm had not yet distinguished the correct answer from the others.

---

## 2. After Oracle

[0.5, 0.5, 0.5, -0.5]

This was the important Grover step.

Only:

|11>

became negative.

So the oracle had encoded the correct answer into:

phase

NOT probability yet.

Initially this felt strange because nothing measurable had changed yet.

That’s because:

(-0.5)^2 = (0.5)^2

So the probabilities were still:

25%

for every state.

The only difference was the hidden phase structure introduced by the oracle.

---

## 3. After Diffusion

[-1.26e-34, -2.36e-17, -9.52e-18, -1.0]

This was the point where interference became visible.

The first three amplitudes became:

almost 0

because destructive interference cancelled them.

Meanwhile:

|11>

became:

-1.0

meaning almost all amplitude had concentrated onto:

|11>

At this point the amplification process finally became much easier to visualize.

Instead of checking states one-by-one, Grover was reshaping amplitudes through interference until the correct answer dominated the probability distribution.

Even though the final amplitude became -1 instead of +1, the probability still became 1 because probabilities depend on |amplitude|^2.

# Search Complexity

Grover’s algorithm instead requires roughly O(sqrt(N)) iterations.

For this example:

N = 4

so:

sqrt(4) = 2

so only a very small number of Grover iterations was enough to strongly amplify the correct state.

As the search space grows larger, this difference becomes much more significant.

# Final Thoughts

Grover’s algorithm was the point where interference started feeling computationally useful rather than just mathematically interesting.

The main realization here was that the oracle itself only introduced a hidden phase difference.

The actual search advantage came from diffusion repeatedly reshaping amplitudes through interference until the correct state dominated the probability distribution.

Seeing the amplitudes directly made the amplification process much easier to understand compared to only looking at final measurement outputs.
