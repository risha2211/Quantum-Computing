# Bloch Sphere Visualization

The Bloch sphere is a geometric way of visualizing the state of a single qubit.

Instead of representing a qubit only algebraically, the quantum state can be viewed as a point on the surface of a sphere.

This makes ideas such as:
- superposition
- phase
- quantum rotations

much easier to interpret visually.

A qubit can exist in infinitely many superposition states.

The Bloch sphere helps visualize these states geometrically.

# Key Locations On The Sphere

North pole: |0>

South pole: |1>

Equator:
superposition states such as:
|+>
|->

Quantum gates can often be interpreted as rotations on the Bloch sphere.

# Why Rotations Matter

Earlier gates such as Ry(theta) mostly felt algebraic.

The Bloch sphere made it easier to see that these gates literally rotate the quantum state around different axes.

This became especially important for:
- variational quantum circuits
- phase intuition
- quantum state evolution

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
import matplotlib.pyplot as plt
import numpy as np

# create circuit
qc = QuantumCircuit(1)

# apply rotation
qc.ry(np.pi/3, 0)

# generate statevector
state = Statevector.from_instruction(qc)

# plot Bloch sphere
bloch = plot_bloch_multivector(state)
display(bloch)
```

# Observation

Initially the qubit started at |0> aka the north pole of the Bloch sphere.

After applying Ry(pi/3) the Bloch vector rotated away from the north pole toward a superposition region of the sphere.

The qubit exists somewhere continuously between the two basis states before measurement occurrs.

The Bloch sphere visualizes the quantum state before measurement.

Once a measurement happens, the state collapses back to |0> or |1> (the poles).

The smooth rotations happen during quantum evolution.

Measurements still produce discrete classical outcomes.

This was probably one of the first times quantum mechanics started feeling visually understandable rather than only mathematical
