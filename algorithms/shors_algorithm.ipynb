# Shor's Algorithm

Shor’s algorithm is a quantum algorithm used for integer factorization.

Given a large number N

we want to find its prime factors.

What made this algorithm important was that it demonstrated that quantum computers could factor integers exponentially faster than known classical algorithms.

# Why This Matters

Modern RSA cryptography depends on the fact that factoring very large integers is computationally difficult for classical computers.

Shor’s algorithm showed that a sufficiently powerful quantum computer could solve this problem much more efficiently.

Shor’s algorithm does not directly factor numbers.

It converts factoring into a problem of finding hidden periodicity.

This is why the algorithm is closely related to Simon’s algorithm.

Both algorithms use interference to detect hidden repeating structure.

# Periodicity Example

Suppose:

N = 15
a = 2

Now consider the function:

f(x) = 2^x mod 15

Evaluating a few values:

x = 0 → 1
x = 1 → 2
x = 2 → 4
x = 3 → 8
x = 4 → 1

At:

x = 4

the outputs begin repeating again.

So the function has period:

r = 4

# Classical vs Quantum Factorization

Classically, factoring a very large number usually involves testing possible factors individually.

For example, a classical computer may repeatedly check:

N ÷ 2
N ÷ 3
N ÷ 4
...

until factors are eventually found.

For extremely large numbers such as 2048-bit RSA integers, this becomes computationally expensive very quickly.

Quantum computing approaches the problem very differently.

Instead of directly searching for factors, Shor’s algorithm studies the function:

f(x) = a^x mod N

for some chosen value of:

a

The important realization is that this function eventually begins repeating periodically.

A quantum computer creates a superposition of many values of:

x

simultaneously and computes:

a^x mod N

for all of them at once.

At first this may seem like enough to immediately solve the problem.

However, directly measuring the quantum state would collapse the superposition and reveal only one random value, destroying the rest of the information.

This is why the algorithm focuses on the behavior of the quantum amplitudes and interference patterns instead of directly measuring intermediate states.

Because the function is periodic, the amplitudes develop repeating wave-like structure.

The Quantum Fourier Transform (QFT) is then used to analyze this interference pattern and extract the hidden period efficiently.

Once the period is known, the factors of:

N

can then be computed using classical mathematics.

# How Period Gives Factors

Once the period:

r

is found, factors can often be computed using:

gcd(a^(r/2) ± 1, N)

For the previous example:

2^(4/2) = 4

Then:

4 - 1 = 3
4 + 1 = 5

which gives the factors:

15 = 3 × 5

# What Quantum Computing Contributes

The difficult part classically is finding the hidden period efficiently.

Shor’s algorithm uses:
- superposition
- interference
- the Quantum Fourier Transform (QFT)

to detect the period much more efficiently than classical algorithms.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.circuit.library import QFTGate

# create circuit
qc = QuantumCircuit(4,4)

# create superposition
qc.h([0,1,2,3])

# apply QFT
qc.append(QFTGate(4), [0,1,2,3])

# measurement
qc.measure([0,1,2,3],[0,1,2,3])

# decompose QFT into basic gates
qc = qc.decompose()

# draw circuit
print(qc.draw())

# run simulator
simulator = AerSimulator()

job = simulator.run(qc, shots=1000)

result = job.result()
counts = result.get_counts()

print(counts)
```

# Observation

This notebook focuses on understanding how the Quantum Fourier Transform manipulates amplitudes and phase relationships.

The circuit first created a superposition of many states simultaneously.

The QFT then applied controlled phase rotations which reshaped the interference pattern of the quantum state.

One important realization was that directly measuring the state too early would collapse the superposition and destroy most of the information.

This is why Shor’s algorithm relies heavily on interference and phase behavior instead of directly observing intermediate quantum states.

The main quantum advantage in Shor’s algorithm does not come from simply computing:

a^x mod N

for many values at once.

The real advantage comes from using interference and the Quantum Fourier Transform to extract hidden periodic structure globally from the wavefunction.
