# Quantum Phase Estimation (QPE)

Quantum Phase Estimation (QPE) is a quantum algorithm used to estimate the phase associated with a quantum state.

The algorithm is foundational in quantum computing because many important problems can ultimately be reduced to phase or eigenvalue estimation.

Suppose there exists:
- a quantum operation:

U

- and a special quantum state:

|ψ>

such that:

U|ψ> = e^(2πiφ)|ψ>

This means applying:

U

does not change the state itself, but instead multiplies it by a phase factor.

The goal of QPE is to estimate:

φ

Phase differences are not directly measurable.

However, phase affects interference patterns, which later affect measurement probabilities.

QPE extracts phase information itself rather than only observing its indirect effects on interference.

The algorithm:
- creates superposition
- applies controlled quantum operations
- encodes phase information into amplitudes
- uses the inverse Quantum Fourier Transform (QFT)
- converts phase information into measurable binary output

A useful way to think about QFT here is as a phase decoder.

# Simple Example

Suppose:

φ = 0.5

Then:

e^(2πiφ)
=
e^(πi)
=
-1

So the quantum operation introduces a phase factor of:

-1

QPE should estimate the phase:

0.5

through measurement.

# Why This Algorithm Is Important

QPE is one of the most important building blocks in quantum computing.

It forms the foundation of:
- Shor’s algorithm
- quantum chemistry algorithms
- Hamiltonian simulation
- many advanced quantum algorithms

This is also one of the first algorithms where the Quantum Fourier Transform becomes central rather than just auxiliary.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.circuit.library import QFTGate
import numpy as np

# create circuit
qc = QuantumCircuit(3,2)

# prepare eigenstate |1>
qc.x(2)

# create superposition on counting qubits
qc.h(0)
qc.h(1)

# controlled phase operations

qc.cp(np.pi, 1, 2)
qc.cp(np.pi/2, 0, 2)

# inverse QFT
qft = QFTGate(2).inverse()

qc.append(qft, [0,1])

# decompose for simulator
qc = qc.decompose()

# measurement
qc.measure([0,1],[0,1])

# draw circuit
print(qc.draw())

# run simulator
simulator = AerSimulator()

job = simulator.run(qc, shots=1000)

result = job.result()
counts = result.get_counts()

print(counts)
```

# Observation

The final measurement result was:

01

for all 1000 shots.

Qiskit displays measurement bits in reversed order.

So, 01 corresponded to 10

0.10₂ = 0.5

This showed that the phase information had been successfully extracted and converted into binary form.

Another important realization was that the phase itself was never measured directly.

Instead:
- controlled phase operations encoded phase information into amplitudes
- interference reshaped the quantum state
- inverse QFT converted the phase relationships into measurable binary output

The algorithm therefore worked by translating hidden phase behavior into readable classical bits.

The inverse Quantum Fourier Transform acted like a phase decoder.

Without inverse QFT, the phase information would remain hidden inside the interference pattern and could not be measured directly.

# Final Thoughts

QPE felt very different from earlier algorithms because the goal was not to search for states or amplify probabilities.

Instead, the algorithm focused on extracting hidden phase information from quantum evolution itself.

This also made the role of phase much more concrete because phase was no longer just indirectly affecting interference, but became the actual quantity being estimated.

The algorithm was especially important because many advanced quantum algorithms ultimately depend on phase estimation internally.
