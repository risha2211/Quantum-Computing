# Deutsch–Jozsa Algorithm

The Deutsch–Jozsa algorithm generalizes the Deutsch algorithm for n-bit Boolean functions.

The goal is to determine whether a function is:
- constant
or
- balanced

using a single oracle query.

# Problem Statement

We have a boolean function f(x)

where:
- x is an n-bit input
- output is either 0 or 1

The function is guaranteed to be either:

1. Constant
   - all outputs are identical

OR

2. Balanced
   - exactly half the outputs are 0
   - exactly half are 1

The goal is to determine whether the function is constant or balanced.

# Classical Complexity

For an n-bit function, there are:

2^n

possible inputs.

Classically, in the worst case, we may need:

2^(n-1) + 1

evaluations to guarantee whether the function is constant or balanced.

# Quantum Advantage

The Deutsch–Jozsa algorithm determines the answer using only:

1 oracle query.

This is achieved using:
- superposition
- phase kickback
- interference

The oracle encodes information into phase, and interference extracts the global property of the function.

The algorithm does not explicitly compute every function output individually.

Instead, phase information introduced by the oracle changes interference patterns across the superposition.

```python
import random

from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# number of input qubits
n = 3

# randomly choose oracle type
oracle_type = random.choice(["constant", "balanced"])

# create circuit
qc = QuantumCircuit(n + 1, n)

# initialize helper qubit to |1>
qc.x(n)

# apply hadamard gates
for qubit in range(n + 1):
    qc.h(qubit)

# oracle section
if oracle_type == "constant":

    # randomly choose:
    # always 0 OR always 1
    constant_output = random.choice([0, 1])

    # if function always returns 1,
    # flip helper qubit
    if constant_output == 1:
        qc.x(n)

else:

    # balanced oracle using CNOT gates
    for qubit in range(n):
        qc.cx(qubit, n)

# apply hadamard again to input qubits
for qubit in range(n):
    qc.h(qubit)

# measure input qubits
for qubit in range(n):
    qc.measure(qubit, qubit)

# draw circuit
print(qc.draw())

# run simulator
simulator = AerSimulator()

job = simulator.run(qc, shots=1000)

result = job.result()
counts = result.get_counts()

print("\nactual oracle type:", oracle_type)
print("measurement results:", counts)

# interpret result
measured_state = list(counts.keys())[0]

if measured_state == "000":
    prediction = "constant"
else:
    prediction = "balanced"

print("algorithm prediction:", prediction)
```

# Output Interpretation

The circuit first creates a superposition of all possible input states.

The oracle then introduces phase changes depending on whether the function is:
- constant
or
- balanced.

After the final Hadamard gates, interference occurs between amplitudes.

If the oracle is constant:
- constructive interference restores the all-zero state
- the final measurement becomes:

000

If the oracle is balanced:
- destructive interference removes the all-zero state
- the final measurement becomes some nonzero bitstring.

Therefore:

- measurement result 000
  → oracle is constant

- any nonzero result
  → oracle is balanced

# Observation

The Deutsch–Jozsa algorithm successfully determines whether the oracle is constant or balanced using a single oracle query.

Instead of checking outputs one-by-one, the algorithm uses:
- superposition
- phase kickback
- interference

to extract a global property of the function.
