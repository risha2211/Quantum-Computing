# Deutsch Algorithm

The Deutsch algorithm determines whether a 1-bit Boolean function is:
- constant
or
- balanced

using a single oracle query.

Deutsch–Jozsa generalizes this idea for n-bit Boolean functions

# Problem Statement

We have a boolean function f(x)
which maps n-bit inputs to 0 or 1.

The function is guaranteed to be either:

1. Constant
   - all outputs are identical

OR

2. Balanced
   - exactly half the outputs are 0
   - exactly half are 1

The goal is to determine whether the function is constant or balanced.

# Oracle Construction

In Deutsch and Deutsch–Jozsa algorithms, the function is implemented as a quantum oracle.

The oracle acts as:

Uf|x,y> = |x, y xor f(x)>

Meaning:
- the first qubit remains unchanged
- the second qubit is modified using XOR with the function output

This construction is used because quantum operations must be reversible.

Examples:

1. Constant function:
   f(x)=0

Then:

y xor 0 = y

So the oracle does nothing.

2. Constant function:
   f(x)=1

Then:

y xor 1

flips the second qubit.

3. Balanced function:
   f(0)=0
   f(1)=1

This behaves like a controlled-NOT operation:
- if x=1, flip the second qubit
- if x=0, leave it unchanged

which can be implemented using a CNOT gate.

# Classical vs Quantum Queries

For a 1-bit Boolean function:

Classically:
- we may need 2 evaluations to determine whether the function is constant or balanced.

Quantum mechanically:
- the Deutsch algorithm determines this using only 1 oracle query.

This is achieved using:
- superposition
- phase
- interference

# Quantum Advantage

The Deutsch–Jozsa algorithm determines the answer using only:

1 oracle query

This is possible because quantum superposition allows amplitudes corresponding to multiple computational basis states to evolve simultaneously.

Interference then extracts global information about the function.

The algorithm then uses interference to combine amplitudes such that:

- balanced functions produce destructive interference
- constant functions produce constructive interference

The final measurement reveals the global property of the function.

# Constant Oracle Example

This example implements a constant function:

f(0)=0
f(1)=0

The oracle therefore performs no operation.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# create 2 qubits & 1 classical bit
qc = QuantumCircuit(2,1)

# initialize second qubit to |1>
qc.x(1)

# apply hadamard to both qubits
qc.h(0)
qc.h(1)

# constant oracle
# does nothing

# apply hadamard again to first qubit
qc.h(0)

# measure first qubit
qc.measure(0,0)

# draw circuit
print(qc.draw())

# run 
simulator = AerSimulator()

job = simulator.run(qc, shots=1000)

result = job.result()
counts = result.get_counts()

print(counts)
```

# Balanced Oracle Example

This example implements a balanced function:

f(0)=0
f(1)=1

This behaves like:

f(x)=x

and is implemented using a CNOT gate.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# create 2 qubits & 1 classical bit
qc = QuantumCircuit(2,1)

# initialize second qubit to |1>
qc.x(1)

# apply hadamard to both qubits
qc.h(0)
qc.h(1)

# balanced oracle
qc.cx(0,1)

# apply hadamard again to first qubit
qc.h(0)

# measure first qubit
qc.measure(0,0)

# draw circuit
print(qc.draw())

# run 
simulator = AerSimulator()

job = simulator.run(qc, shots=1000)

result = job.result()
counts = result.get_counts()

print(counts)
```

# Comparison

| Oracle Type | Measurement Result | Interpretation |
|---|---|---|
| Constant | 0 | Function is constant |
| Balanced | 1 | Function is balanced |

The difference arises from interference patterns introduced by the oracle.
