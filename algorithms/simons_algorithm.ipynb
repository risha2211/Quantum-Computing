# Simon's Algorithm

Simon’s algorithm is used to discover a hidden binary string encoded inside a special type of function.

What made this algorithm interesting was that the measurement results did not directly reveal the answer.

Instead, each measurement produced partial information about the hidden structure, and combining multiple results eventually revealed the secret string.

# The Main Idea

Suppose there exists a hidden binary string: s

Example: 101

We are given access to a function f(x)

with a special property: f(x) = f(y)

only when: x ⊕ y = s

(⊕ = represents bitwise XOR)

# Why Classical Computing Struggles

Classically, discovering the hidden string requires finding two different inputs that produce the same output.

As the number of bits grows, this becomes exponentially difficult because many function queries may be needed before a matching pair is found.

# What Quantum Computing Does Differently

Simon’s algorithm creates a superposition of many inputs at once.

After interference and measurement, the algorithm does not directly reveal the hidden string.

Instead, each measurement produces a constraint of the form:

z · s = 0 mod 2

where:
- z is the measurement result
- s is the hidden string
- · represents bitwise dot product

Running the algorithm multiple times produces multiple equations which can then be solved to recover the hidden string.

# Why This Algorithm Matters

Simon’s algorithm was historically important because it demonstrated one of the first exponential quantum speedups.

It also introduced the idea that quantum interference can reveal hidden periodic structure (outputs repeat for inputs separated by s).

This idea later became foundational for Shor’s factoring algorithm.

# Important Shift In Intuition

In algorithms like Bernstein–Vazirani, the final measurement directly revealed the hidden string.

Simon’s algorithm felt different because the measurements only revealed partial constraints.

The hidden string is reconstructed gradually by combining information from multiple runs.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# create circuit
qc = QuantumCircuit(4,2)

# apply Hadamard gates
# to input qubits
qc.h(0)
qc.h(1)

# oracle implementation
# hidden string s = 11

qc.cx(0,2)
qc.cx(1,2)

qc.cx(0,3)
qc.cx(1,3)

# apply Hadamard again
qc.h(0)
qc.h(1)

# measure input qubits
qc.measure(0,0)
qc.measure(1,1)

# draw circuit
print(qc.draw())

# run simulator
simulator = AerSimulator()

job = simulator.run(qc, shots=1000)

result = job.result()
counts = result.get_counts()

print(counts)
```

# Observation

Unlike algorithms such as Bernstein–Vazirani or Grover, the measurement results in Simon’s algorithm did not directly reveal the hidden string.

Instead, the outputs:

00
11

represented constraint vectors satisfying:

z · s = 0 mod 2

The important result was not which output appeared more often, but the fact that only certain outputs survived after interference while invalid states were cancelled out.

This made the role of interference feel very different from Grover’s algorithm.

Instead of amplifying one correct answer, Simon’s algorithm used interference to filter out states incompatible with the hidden structure.

After interference, the final quantum state still remained in superposition.

This is why the 1000 shots were distributed roughly equally between 00 & 11

rather than collapsing to only one result.

The key point was that both states satisfied the constraint equation, so both remained measurable with nonzero probability.
