# Quantum Circuit Training Loop

Earlier notebooks introduced:
- variational quantum circuits
- trainable rotation parameters
- quantum classifier structures

However, the parameters were still being chosen manually.

The important missing piece was that the circuit itself was not yet learning from mistakes.

This notebook introduces the optimization loop that allows quantum model parameters to gradually improve automatically through repeated updates.

# What Makes This ML

Earlier variational quantum circuits mainly demonstrated that:
- quantum circuits can contain trainable parameters
- changing rotation angles changes measurement probabilities

However, true machine learning also requires:
- predictions
- error evaluation
- parameter updates
- iterative improvement

We introduce that feedback loop.

The circuit will:
- make a prediction
- compare it with a target value
- compute error
- adjust parameters
- repeat the process

The parameter gradually moves toward values that produce better predictions.

This is similar to optimization in classical machine learning systems.

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
import numpy as np

# target probability
target = 0.8

# initial parameter
theta = 0.1

# learning rate
lr = 0.1

# training loop
for step in range(20):

    # create circuit
    qc = QuantumCircuit(1)

    # trainable rotation
    qc.ry(theta, 0)

    # get statevector
    state = Statevector.from_instruction(qc)

    # probability of measuring 1
    prob_1 = abs(state.data[1])**2

    # compute error
    loss = (target - prob_1)**2

    # print progress
    print(f"Step {step}")
    print(f"Theta = {theta:.4f}")
    print(f"Probability of 1 = {prob_1:.4f}")
    print(f"Loss = {loss:.6f}")
    print()

    # simple manual update rule
    if prob_1 < target:
        theta += lr
    else:
        theta -= lr
```

# Observation

The training process began with theta = 0.1

At this stage, the qubit only rotated slightly away from |0> on the Bloch sphere.

Because of this, the probability of measuring 1 was extremely small (0.0025), very far from the target probability of 0.8 so the loss initially became quite large.

The optimization loop then repeatedly checked whether the prediction was too low or too high.

Since the probability of measuring 1 was still below the target, the circuit gradually increased theta during every iteration.

As theta increased, the qubit rotated farther away from |0>, toward |1> on the Bloch sphere increasing the probability of measuring 1.

The circuit was no longer behaving like a fixed static program but learned from feedback and changed automatically in response to prediction error.
