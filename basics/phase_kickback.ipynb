# Phase Kickback

Phase kickback is a phenomenon in quantum computing where information about a function becomes encoded as phase on a quantum state.

Instead of storing function outputs directly, quantum algorithms often store information using phase differences.

These phase differences later affect interference patterns and final measurement outcomes.

Classical algorithms usually compute outputs one-by-one.

Quantum algorithms instead manipulate:
- amplitudes
- phase
- interference

Many important quantum algorithms use phase kickback, including:
- Deutsch algorithm
- Deutsch–Jozsa algorithm
- Grover’s algorithm
- Shor’s algorithm

# Oracle Definition

In many quantum algorithms, a function is implemented using a quantum oracle.

The oracle acts as:

Uf|x,y> = |x, y xor f(x)>

Meaning:
- the first qubit remains unchanged
- the second qubit is modified using XOR with the function output

This construction is used because quantum operations must be reversible.

# Preparing the Target Qubit

Suppose the second qubit is prepared as:

(|0> - |1>)/sqrt(2)

This state is important because flipping the qubit introduces a negative phase.

For example:

X(|0> - |1>) = (|1> - |0>) = -(|0> - |1>)

The qubit returns to the same form but gains a negative sign.

# Phase Kickback Intuition

Suppose:

f(x)=x

If:
- x=0, the oracle does nothing
- x=1, the oracle flips the target qubit

When the target qubit is in the state:

(|0> - |1>)/sqrt(2)

the flip introduces a negative phase.

As a result:

|x>

picks up a phase factor:

(-1)^f(x)

The function information is therefore encoded as phase.

# General Form

A quantum superposition:

Σ |x>

transforms into:

Σ (-1)^f(x)|x>

The amplitudes remain present, but some basis states acquire phase differences depending on the function output.

Two quantum states may produce identical measurement probabilities while still having different phases.

These hidden phase differences later affect interference.

Quantum algorithms exploit this interference to:
- amplify useful answers
- suppress incorrect answers

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

qc = QuantumCircuit(2,1)

# prepare target qubit as |1>
qc.x(1)

# create superposition and phase state
qc.h(0)
qc.h(1)

# oracle
qc.cx(0,1)

# interference
qc.h(0)

# measurement
qc.measure(0,0)

print(qc.draw())

simulator = AerSimulator()

job = simulator.run(qc, shots=1000)

result = job.result()
counts = result.get_counts()

print(counts)
```

# Observation

The oracle does not directly modify the first qubit.

However, phase information introduced through the target qubit changes the interference pattern of the first qubit.

This phenomenon is called phase kickback.

# Output Analysis

The final measurement result is:

{'1': 1000}

This means:
- the first qubit was measured as 1 in every run
- the interference pattern became fully constructive for |1>
- the |0> amplitude cancelled through destructive interference

The oracle itself only acted on the second qubit.

However, because the target qubit was prepared in the state:

(|0> - |1>)/sqrt(2)

the oracle introduced phase information into the amplitudes of the first qubit.

This hidden phase difference later changed the interference outcome after the final Hadamard gate.

As a result:
- information about the function became encoded as phase
- the measurement of the first qubit revealed this information

This phenomenon is called phase kickback.

# Key Insight

Quantum algorithms often do not store useful information directly as classical values.

Instead, information is encoded into:
- amplitudes
- relative phase

Interference then converts these hidden phase differences into measurable outcomes.
