# Superposition using Hadamard Gate

Initial state : |0>

Hadamard gate transforms:

|0> => (|0> + |1>)/sqrt(2)

This creates an equal superposition of states 0 and 1.

After measurement:
Probability of 0 = 50%
Probability of 1 = 50%

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# Create a circuit with:
# 1 quantum bit
# 1 classical bit
qc = QuantumCircuit(1, 1)

# Apply Hadamard gate
qc.h(0)

# Measure the qubit
qc.measure(0, 0)

# Draw the circuit
print(qc.draw())

# Create simulator
simulator = AerSimulator()

# Run the circuit 1000 times
job = simulator.run(qc, shots=1000)

# Get results
result = job.result()
counts = result.get_counts()

print(counts)
```

Observation:

The output distribution is approximately 50–50 between 0 and 1 after many runs.

Initially this seemed similar to random.randint(0,1), but the key difference is that quantum systems evolve using probability amplitudes, which can later interfere with each other.
