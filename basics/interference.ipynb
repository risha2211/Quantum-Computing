# Quantum Interference

This notebook explores quantum interference using:
- Hadamard gate
- Z gate
- phase changes

The experiment demonstrates how hidden phase differences can change final measurement outcomes through constructive and destructive interference.

# Circuit Logic

Initial state:

|0>

Step 1:
Apply Hadamard:

|0> => (|0> + |1>)/sqrt(2)

Step 2:
Apply Z gate:

(|0> + |1>)/sqrt(2) => (|0> - |1>)/sqrt(2)

Step 3:
Apply Hadamard again.

The final interference pattern determines the measurement outcome.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

qc = QuantumCircuit(1,1)

# create superposition
qc.h(0)

# introduce phase change
qc.z(0)

# Hadamard
qc.h(0)

qc.measure(0,0)

# draw circuit
print(qc.draw())

# run
simulator = AerSimulator()
job = simulator.run(qc, shots=1000)

result = job.result()
counts = result.get_counts()

print(counts)
```

# Observation

The final measurement becomes 1 with nearly* 100% probability.

The Z gate did not directly change measurement probabilities when applied alone.

However, it introduced a phase difference that later affected interference after the second Hadamard gate.

This demonstrates how quantum computation manipulates amplitudes and phase rather than classical probabilities alone.

*nearly because in real quantum hardware we find noise, decoherence, gates are imperfect, measurements have errors.
