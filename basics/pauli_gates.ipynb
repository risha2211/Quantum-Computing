# Quantum Gates Playground

This notebook explores the behavior of basic single-qubit quantum gates:
- Pauli-X gate
- Pauli-Y gate
- Pauli-Z gate

The goal is to build a foundation for:
- state transformations
- superposition
- phase
- interference

# X Gate

The X gate acts similarly to a classical NOT gate.

Matrix representation:

X = [[0,1],
     [1,0]]

Transformations:

|0> => |1⟩
|1> => |0⟩

X does not create superposition.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# create 1 qubit and 1 classical bit
qc = QuantumCircuit(1,1)

# apply Pauli-X gate
qc.x(0)

qc.measure(0,0)

print(qc.draw())

# run
simulator = AerSimulator()
job = simulator.run(qc, shots=1000)

result = job.result()
counts = result.get_counts()

print(counts)
```

Observation:

The measurement result is always 1 because the X gate deterministically flips |0> to |1>

# Z Gate

The Z gate introduces phase.

Matrix representation:

Z = [[1,0],
     [0,-1]]

Transformations:

|0> => |0>
|1> => -|1>

The Z gate changes the sign (phase) of the |1> amplitude without directly changing measurement probabilities.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

qc = QuantumCircuit(1,1)

qc.h(0)
qc.z(0)
qc.measure(0,0)

print(qc.draw())

simulator = AerSimulator()
job = simulator.run(qc, shots=1000)

result = job.result()
counts = result.get_counts()

print(counts)
```

Observation:

The probabilities remain approximately 50–50 after applying H followed by Z.

However, the internal quantum state changes from:

(|0> + |1>)/sqrt(2)

to:

(|0⟩ - |1⟩)/sqrt(2)

This phase difference later affects interference behavior.

# Y Gate

The Y gate combines bit-flip behavior with phase changes.

Matrix representation:

Y = [[0,-i],
     [i,0]]

Transformations:

|0> => i|1>
|1> => -i|0>

This introduces complex amplitudes into the quantum state.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

qc = QuantumCircuit(1,1)

qc.y(0)

qc.measure(0,0)

print(qc.draw())

simulator = AerSimulator()
job = simulator.run(qc, shots=1000)

result = job.result()
counts = result.get_counts()

print(counts)
```

Observation:

Although the Y gate introduces imaginary phase factors, measurement probabilities remain deterministic in this experiment because the resulting state is still fully |1> up to a phase factor.

# Key Insight

Quantum gates do more than flip bits.

They manipulate:
- amplitudes
- phases
- interference patterns

Two quantum states can produce identical measurement probabilities while still being fundamentally different internally due to phase.
