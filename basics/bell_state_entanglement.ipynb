# Bell State and Quantum Entanglement

Initial state:

|00>

Applying Hadamard to the first qubit gives:

|00> => (|00> + |10>)/sqrt(2)

Applying CNOT creates the Bell state:

(|00> + |11>)/sqrt(2)

This creates entanglement between the two qubits.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# create 2 qubits & 2 classical bits
qc = QuantumCircuit(2, 2)

# put first qubit into superposition
qc.h(0)

# CNOT: if q0 becomes 1, flip q1
qc.cx(0, 1)

# measure both qubits
qc.measure([0, 1], [0, 1])

# display circuit
print(qc.draw())

# simulator
simulator = AerSimulator()

# run
job = simulator.run(qc, shots=1000)

# results
result = job.result()
counts = result.get_counts()

print(counts)
```

Observation:

Only:
00
11

appear in the measurement results.

The states:
01
10

do not appear because the qubits are entangled and their measurements become correlated.
