# Variational Quantum Circuits (VQC)

Variational Quantum Circuits (VQCs) are quantum circuits containing trainable parameters.

VQCs can adjust their behavior by changing gate parameters. This makes the circuit itself trainable.

VQCs use parameterized gates like:

Ry(θ) where θ is a tunable parameter.

Changing θ changes:
- amplitudes
- interference patterns
- measurement probabilities

So the quantum circuit can gradually learn useful behavior through optimization.

In classical machine learning:
- models contain trainable weights
- a loss function is computed
- optimization adjusts the weights

Variational quantum circuits work similarly
- gate parameters act like trainable weights
- measurements produce outputs
- a classical optimizer updates the parameters

# Hybrid Quantum-Classical Computation

In VQCs the computation is hybrid.

The quantum computer:
- prepares states
- creates interference
- performs measurements

The classical computer:
- computes the loss
- updates parameters
- decides how the circuit should change

So both systems work together during training.

# Why Variational Circuits Became Popular

Large quantum algorithms such as Shor’s algorithm often require extremely deep and precise quantum circuits.

The problem is that current quantum hardware is still very noisy and unstable.

Every additional gate introduces:
- more noise
- more decoherence
- higher chances of destroying interference patterns

This makes large algorithms difficult to run reliably on present-day quantum devices.

Variational circuits became important because they usually use:
- shorter circuits
- fewer gates
- shallower circuit depth

which makes them much more practical for current hardware.

Also, VQCs are optimization-driven rather than requiring one perfectly exact computation.

The circuit gradually adjusts its parameters over multiple runs.

This makes VQCs more tolerant to hardware imperfections because the optimizer can keep adapting the parameters experimentally based on measurement results.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
import numpy as np

# parameter
theta = np.pi / 4

# create circuit
qc = QuantumCircuit(1,1)

# parameterized rotation gate
qc.ry(theta, 0)

# measurement
qc.measure(0,0)

# draw circuit
print(qc.draw())

# run simulator
simulator = AerSimulator()

job = simulator.run(qc, shots=1000)

result = job.result()
counts = result.get_counts()

print(counts)
```

# Observation

The qubit initially started in:

|0>

The:

Ry(π/4)

gate then rotated the state partially toward:

|1>

instead of flipping it completely.

This created a superposition where:
- the amplitude for |0> remained larger
- the amplitude for |1> became smaller but nonzero

After measurement, the output became:

{'0': 856, '1': 144} (first time)

This showed that the qubit now had:
- a higher probability of collapsing to 0
- a smaller probability of collapsing to 1

The most important realization here was that changing the parameter:

θ

directly changed the measurement probabilities.

So the circuit behavior itself became tunable and trainable.

This was the first notebook where the quantum circuit started behaving like an adaptive system instead of a fixed algorithm.

Rather than searching for one exact answer, the circuit adjust probabilities.

This made quantum computation feel much closer to a learning process rather than only a deterministic computation.
