# Quantum Classifier

A quantum classifier is a simple quantum machine learning model used to separate or classify data.

Instead of using only classical mathematical transformations, the model uses:
- quantum states
- parameterized gates
- interference
- measurement probabilities

to produce predictions.

The basic workflow becomes:

classical data
- encode into qubits
- apply trainable quantum circuit
- measure output
- make prediction

The trainable parameters inside the circuit are adjusted over time so that the predictions gradually improve.

# Data Encoding

We convert normal classical data into quantum states.

E.g.: Using rotation gates
Ry(x) can encode a classical value x into the amplitudes of a qubit state.

Unlike classical neural networks which directly output numbers, quantum circuits output probabilities.

So after measurement:
- the probability of measuring 0
- the probability of measuring 1

can be interpreted as predictions.

# Hybrid Learning

The quantum circuit itself performs:
- superposition
- interference
- state evolution

while a classical optimizer updates the trainable parameters.

So the learning process becomes hybrid:
- quantum computation
- classical optimization
working together repeatedly.

Earlier algorithms tried to find exact hidden answers.

Quantum classifiers gradually reshape probability distributions.

The model slowly learns by adjusting how amplitudes interfere with each other.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
import numpy as np

# input data
x = np.pi / 3

# trainable parameter
theta = np.pi / 4

# create circuit
qc = QuantumCircuit(1,1)

# encode classical data
qc.ry(x, 0)

# trainable layer
qc.ry(theta, 0)

# measurement
qc.measure(0,0)

# draw circuit
print(qc.draw())

# run simulator
simulator = AerSimulator()

job = simulator.run(qc, shots=1000)

result = job.result()
counts = result.get_counts()

print(counts)
```

The first Ry(x) gate encodes the input data into the qubit state.

The second Ry(theta) gate acts like a trainable parameter layer.

Changing θ changes how the encoded data evolves before measurement.

The final probabilities can then be interpreted as the model’s prediction.

# Understanding the Output

The circuit first encoded the classical input x = π/3 into the qubit using Ry(x)

This rotated the qubit away from the pure |0> state and stored the input information inside the quantum amplitudes.

After this, the circuit applied another rotation Ry(theta), where θ = π/4

This second rotation acted like a trainable parameter layer.

The circuit was no longer fixed.

Changing theta would directly change:
- amplitudes
- interference behavior
- measurement probabilities

which meant the circuit itself became trainable.

After measurement, the output became:

{'0': 367, '1': 633} (first time)

This meant:
- the qubit collapsed to 0 around 36% of the time
- the qubit collapsed to 1 around 63% of the time

So after both rotations, the final quantum state ended up closer to |1> than |0>

Another important realization was that these probabilities could now be interpreted as predictions.

For example:
- higher probability of 1 - model leans toward class 1
- higher probability of 0 - model leans toward class 0

This was the first notebook where the quantum circuit started behaving like a probabilistic decision-making system rather than only solving a fixed mathematical problem.
