# BB84 Quantum Key Distribution

Most modern cryptography today, such as RSA, depends on mathematical problems being difficult to solve computationally.

BB84 is fundamentally different.

Its security does not come from hard mathematics alone, but from the laws of quantum mechanics.

In classical communication, someone can potentially copy information without being noticed.

In quantum communication, measuring the information itself changes the quantum state.

So instead of being protected only by computational difficulty, the communication is protected by physics.

# Main Idea

Alice wants to send a secret key to Bob.

To do this, she sends qubits prepared using random bases.

Bob also measures using randomly chosen bases.

If Bob uses the correct basis, he recovers the correct bit.

If he uses the wrong basis, the result becomes random.

After transmission:
- Alice and Bob publicly compare which bases they used
- bits measured using matching bases are kept
- the remaining bits are discarded

These remaining bits form the secret key.

Suppose Alice wants to send the bits: 1 0 1 1

She randomly chooses bases: + × + ×

Bob does not know which bases Alice used, so he randomly chooses his own bases: + + × ×

Now compare the bases position-by-position:

1st position:
+ and +  → match

2nd position:
× and +  → mismatch

3rd position:
+ and ×  → mismatch

4th position:
× and ×  → match

Only positions with matching bases are kept because those measurements are reliable.

So positions 1 and 4 are kept, while the others are discarded.

These remaining bits become part of the secret key.

In real quantum communication systems, these bases are implemented physically.

For example, in photon-based systems:
- one basis may use horizontal and vertical polarization
- another may use diagonal polarizations such as 45° and 135°

The important requirement is that the bases must be incompatible, so measuring in the wrong basis disturbs the quantum state.

# Why Eavesdropping Can Be Detected

A key idea in quantum mechanics is that measurement disturbs the state.

If an eavesdropper tries measuring the qubits:
- the quantum states may collapse incorrectly
- this introduces errors into the transmission

Alice and Bob can detect this by comparing a small subset of their bits publicly.

If the error rate is unusually high, they know someone may have intercepted the communication.

# Basis States Used

BB84 uses two bases.

Computational basis (+):

|0>
|1>

Hadamard basis (×):

|+> = (|0> + |1>)/sqrt(2)

|-> = (|0> - |1>)/sqrt(2)

Using random bases is what makes eavesdropping detectable.

```python
import random

# number of bits
n = 10

# alice generates random bits

alice_bits = [
    random.choice([0,1])
    for _ in range(n)
]

# alice randomly chooses bases: + or x

alice_bases = [
    random.choice(["+","x"])
    for _ in range(n)
]

# bob randomly chooses bases

bob_bases = [
    random.choice(["+","x"])
    for _ in range(n)
]

# bob measures qubits

bob_results = []

for i in range(n):

    # if bases match,
    # bob gets correct bit
    if alice_bases[i] == bob_bases[i]:

        bob_results.append(
            alice_bits[i]
        )

    # otherwise result is random
    else:

        bob_results.append(
            random.choice([0,1])
        )

# keep only matching bases

shared_key = []

for i in range(n):

    if alice_bases[i] == bob_bases[i]:

        shared_key.append(
            alice_bits[i]
        )

# display results

print("alice bits:      ", alice_bits)
print("alice bases:     ", alice_bases)

print()

print("bob bases:       ", bob_bases)
print("bob measurements:", bob_results)

print()

print("shared secret key:", shared_key)
```

# Observation

Alice and Bob initially chose their bases independently and randomly.

Because of this:
- some basis choices matched
- some did not

When the bases matched, Bob recovered the correct bit reliably.

When the bases did not match, the measurement outcome became random, so those bits were discarded.

This means BB84 does lose part of the transmitted data, but the tradeoff is improved security.

The important idea is that even though many bits are discarded, the remaining matching bits can still be used to form a secure shared key.

# Important Insight

BB84 is mainly used for secure key distribution rather than transmitting large amounts of data directly.

Once Alice and Bob successfully establish a secure shared key, they can use classical encryption methods like AES to securely exchange actual information.

So the expensive quantum part is mainly used to create the secure key itself.
