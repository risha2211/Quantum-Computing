# Post-Quantum Cryptography (PQC)

Post-Quantum Cryptography (PQC) refers to cryptographic systems designed to remain secure even against quantum computers.

This became important because algorithms such as Shor’s algorithm showed that sufficiently powerful quantum computers could break many classical public-key cryptosystems.

# Why RSA Becomes Vulnerable

Classical RSA security depends heavily on the difficulty of factoring very large integers.

Classically, factoring huge numbers is computationally expensive.

But Shor’s algorithm showed that quantum computers could factor these numbers exponentially faster.

This means that systems such as:
- RSA
- elliptic curve cryptography (ECC)

become vulnerable once large-scale fault-tolerant quantum computers exist.

Initially it felt like quantum computing would destroy modern cryptography completely but in reality only certain cryptographic systems become vulnerable.

For example:
- RSA and ECC are heavily threatened
- symmetric cryptography survives much better

Algorithms such as AES mainly become weaker through Grover’s algorithm rather than completely broken.

# What "Quantum-Safe" Means

A post-quantum cryptographic system is considered quantum-safe if no efficient quantum attack is currently known against it.

The goal is to design cryptographic problems that remain computationally difficult even for quantum computers.

# Lattice Cryptography

One of the biggest directions in post-quantum cryptography is lattice-based cryptography.

Instead of relying on:
- factoring
- discrete logarithms

these systems rely on difficult geometric problems involving very high-dimensional lattices.

The important realization was that quantum computers currently do not have efficient algorithms for solving these lattice problems the way Shor solves factoring.

# Kyber and Dilithium

Modern post-quantum cryptography heavily uses systems such as:
- Kyber
- Dilithium

These were standardized by NIST as important post-quantum cryptographic algorithms.

Kyber is mainly used for:
- key exchange
- encryption

Dilithium is mainly used for:
- digital signatures
- authentication

# Harvest Now, Decrypt Later

One important cybersecurity concern is: "harvest now, decrypt later"

This means attackers may steal encrypted data today and simply store it until sufficiently powerful quantum computers exist in the future.

Even if the data cannot be decrypted now, it may become vulnerable later once quantum attacks become practical.

Earlier in quantum computing, quantum algorithms mostly felt like offensive tools against cryptography.

Post-quantum cryptography introduced the opposite perspective.

Instead of cryptography collapsing completely, cryptographic systems themselves began evolving to survive the quantum era.

```python
# toy lattice code

import numpy as np
import matplotlib.pyplot as plt

# basis vectors
v1 = np.array([2,0])
v2 = np.array([1,2])

# generate lattice points
points = []

for a in range(-5,6):
    for b in range(-5,6):
        point = a*v1 + b*v2
        points.append(point)

points = np.array(points)

# plot
plt.figure(figsize=(6,6))

plt.scatter(points[:,0], points[:,1])

# draw basis vectors
plt.arrow(0,0,v1[0],v1[1], head_width=0.2, length_includes_head=True)
plt.arrow(0,0,v2[0],v2[1], head_width=0.2, length_includes_head=True)

plt.grid(True)
plt.axis('equal')

plt.title("Toy Lattice Structure")

plt.show()
```

# Understanding The Lattice

The two basis vectors:

v₁ = (2,0)
v₂ = (1,2)

generated an entire repeating structure of points.

Every lattice point was formed using integer combinations of the basis vectors:

a·v₁ + b·v₂

The lattice just looks like a tilted grid.

But post-quantum cryptography does not use tiny 2D lattices like this.

Real lattice cryptography uses extremely high-dimensional lattices containing huge numbers of possible points and geometric relationships.

That is where the computational difficulty comes from.

Instead of relying on:
- factoring
- divisibility
- prime numbers

lattice cryptography relies on extremely difficult geometric problems in very high-dimensional spaces.

```python
import math

# classical brute force complexity
classical_128 = 2**128

# grover complexity
quantum_128 = 2**64

print("Classical brute force for 128-bit key:")
print(classical_128)

print("\nApproximate Grover complexity:")
print(quantum_128)

print("\nKey insight:")
print("Grover roughly square-roots brute force search complexity.")

# doubling key size
classical_256 = 2**256
quantum_256 = 2**128

print("\nFor 256-bit keys:")

print("\nClassical complexity:")
print(classical_256)

print("\nGrover complexity:")
print(quantum_256)
```

# Observation

For a 128-bit key classical brute force required roughly 2^128 operations.

But with Grover’s algorithm, the effective search complexity became approximately 2^64

Quantum computers did not completely destroy symmetric cryptography the way Shor threatens RSA.

Grover produced a quadratic speedup.

Increasing key sizes still works as a defense.

For example: 256-bit keys under Grover still behave roughly like 128-bit security classically, which is still extremely strong.
