# Quantum Random Number Generator (QRNG)

Randomness is extremely important in cybersecurity and cryptography.

Encryption systems depend heavily on unpredictable random values for:
- secret keys
- authentication tokens
- secure communication

If the randomness becomes predictable, the security of the system can collapse.

# Classical Randomness vs Quantum Randomness

Most classical computers use pseudo-random number generators (PRNGs).

These generators produce sequences that appear random but are ultimately generated by deterministic algorithms.

Quantum mechanics is different because measurement outcomes can be fundamentally unpredictable.

This allows quantum systems to generate true randomness rather than only simulated randomness.

# Why Classical Randomness Is Usually Not Truly Random

Initially it felt confusing because functions like:

random.randint()

already seem random.

But the important realization was that most classical computers use pseudo-random number generators (PRNGs).

These generators do not create true randomness.

Instead, they use:
- mathematical formulas
- internal states
- seed values

to generate sequences that only appear random.

This means that if someone knows:
- the algorithm
- the initial seed

then the future outputs can theoretically be predicted exactly.

So classical randomness is usually:
- computationally unpredictable
- but still fundamentally deterministic underneath.

# What About Physical Randomness?

Another doubt was whether physical randomness like:
- dice rolls
- coin tosses
- human choices

could count as true randomness.

The interesting realization was that classical physics is usually considered deterministic.

So in principle, if someone somehow knew:
- every force
- every particle position
- every velocity

then even things like dice rolls could theoretically become predictable.

The randomness mostly comes from practical complexity rather than fundamental uncertainty.

Human thinking also initially felt like it might be truly random.

But human decisions are still influenced by:
- brain states
- memories
- subconscious patterns
- physical processes

So human choices are usually considered extremely complicated forms of unpredictability rather than fundamentally random processes.

# Why Quantum Randomness Is Different

Quantum mechanics introduces a very different kind of randomness.

Suppose a qubit is prepared in the superposition state:

|+>

which equals:

(|0> + |1>) / √2

When measured:
- the qubit collapses to 0 with 50% probability
- the qubit collapses to 1 with 50% probability

The important realization was that the outcome is not believed to be secretly predetermined beforehand in standard quantum mechanics.

Even if someone knew everything physically possible about the system, the exact measurement result would still remain fundamentally probabilistic.

This is what makes quantum randomness very different from classical unpredictability.

# Why This Matters For Cybersecurity

Quantum randomness is extremely valuable because nobody can perfectly predict future outputs.

This makes QRNG useful for:
- encryption keys
- authentication systems
- secure communication
- cryptographic protocols

Earlier in quantum computing, measurement randomness often felt annoying because it destroyed information.

In QRNG, that same randomness becomes the useful resource itself

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# create circuit
qc = QuantumCircuit(1,1)

# create superposition
qc.h(0)

# measure qubit
qc.measure(0,0)

# draw circuit
print(qc.draw())

# run
simulator = AerSimulator()

job = simulator.run(qc, shots=1000)

result = job.result()
counts = result.get_counts()

print(counts)

# generate random bit string
random_bits = ""

for outcome, count in counts.items():
    random_bits += outcome * count

print("\nGenerated Random Bits:")
print(random_bits[:50])
```

# Observation

The Hadamard gate first created the superposition state:

|+>

which contained equal amplitudes for:
- |0>
- |1>

After measurement, the qubit randomly collapsed into either:
- 0
- 1

Running the circuit for 1000 shots produced an output close to:

50/50

which matched the expected probability distribution.

The generated bit string looked random, but the interesting realization was that unlike classical pseudo-random generators, the randomness here came directly from quantum measurement itself.
